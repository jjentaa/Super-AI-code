```python
def remove_space(text):
  text = text.strip()
  return text
```

```python
def remove_lakana(text):
  sth = text.split('และคณะ')[0]
  ans = remove_space(sth)
  return ans
```

```python
def to_title(text):
  text = text.title()
  text = remove_space(text)
  return text
```

```python
def remove_special_sym(text):
  text = text.replace('*' , '')
  text = text.replace('\"' , '')
  text = text.replace(',' , '')
  text = remove_space(text)

  return text
```

```python
def only_one_space(text):
  name = list(text)
  res = []
  res2 = []

  for i in range(0 , len(name)) :

    if name[i] == ' ' and name[i - 1] == ' ' and i > 0 :
      res.append('!')
    else :
      res.append(name[i])

  for i in res :
    if i != '!' :
      res2.append(i)

  ans = ''
  for i in res2 :
    ans += i

  ans = remove_space(ans)

  return ans
```

```python
def remove_suffix(text):
  if ('(ผู้แต่ง)' in text and text!=None):
    text = text.replace('(ผู้แต่ง)' , '')
    
  if ('Et.Al' in text and text!=None):
    text = text.replace('Et.Al' , '')
  text = remove_space(text)
  return text
```

```python
def remove_dash(text) :
  if ('-' in text):
    text = list(text)
    res = ''
    res += text[0]

    for i in range(1 , len(text) - 1) :
      if ((text[i] == '-') and (text[i - 1] == ' ') and (text[i + 1] == ' ')):
          pass
      else :
          res += text[i]

      res += text[-1]

      text = res


  if (('-' in text) and (text[-1]=='-')):
    text = text.replace('-', ' ' , 1)

  text = only_one_space(text)
  text = remove_space(text)

  return text
```

```python
def check_language(text):
  english_words = re.findall(r'[A-Z\.a-zÀ-ÖØ-öø-ÿ-]+', text)
  thai_words = re.findall(r'[\u0E00-\u0E7F]+', text)
  chinese_words = re.findall(r'[\u4e00-\u9fff]+', text)
  arabic_words = re.findall(r'/[\u0600-\u065F\u066A-\u06EF\u06FA-\u06FF]/+', text)


  thai_words = ' '.join(thai_words)
  english_words = ' '.join(english_words)
  chinese_words = ' '.join(chinese_words)
  arabic_words = ' '.join(arabic_words)

  return [thai_words, english_words, chinese_words]
```

```python
def check_phra(text):
  if('พระ' in text and text[0:2]=='พระ'):
    text = text.replace('พระ', '')
  elif('Phra' in text and text[0:3]=='Phra'):
    text = text.replace('Phra', '')
  
  text = remove_space(text)
  return text
```

```python
def special_case(text):
  if str(text) != "" and str(text) != None :
    text = text.replace('พระครู', '')
    text = text.replace('Phrakru', '')
    text = text.replace('พระมหา', '')
    text = text.replace('JPHD', '')
    text = text.replace('Jphd', '')
    text = text.replace('พระจำลอง สุภจาโร', 'จำลอง สุภจาโร')
    text = text.replace('วรวรรณ ตุมมงคล', 'วรวรรณ ตุมมงคล')
    text = text.replace('เสาวลักษณ์ กูเจริญประสิทธิ์', 'เสาวลักษณ์ กูเจริญประสิทธิ์')
    text = text.replace('ว่าที่ ร.ต. สุวุฒิ สุกิจจากร', 'สุวุฒิ สุกิจจากร')
    text = text.replace('ืnatthabhol khanthachai', 'natthabhol khanthachai')
    text = text.replace('สมชัย วงษ์ยะ', 'สมชัย วงษ์นายะ')
    text = text.replace('อดุลย์ กุสลจิตฺโต(ภักดีกุล)', 'อดุลย์ กุสลจิตฺโต')
    text = text.replace('พัฒนาตำราและ วิทยาลัยดุสิตธานี', '')
    text = text.replace('Dusit Thani College', '')
    text = text.replace('รัฏฐาภิรักษ์', '')
    #text = text.replace('Ha', '')
    text = text.replace('อดุลย์ กุสลจิตฺโต ภักดีกุล', 'อดุลย์ กุสลจิตฺโต(ภักดีกุล)')
    text = text.replace('อิทธิเดช เพ็ชร', 'อิทธิเดช พระเพ็ชร')
    text = text.replace('Cfa', '')
    text = remove_space(text)

  return text
```

```python
def cleaner(text):
  text = remove_space(text)
  text = remove_prefix_status(text)
  text = remove_special_sym(text)
  text = only_one_space(text)
  text = remove_lakana(text)
  text = remove_dash(text)
  text = remove_suffix(text)
  text = to_title(text)
  text = check_language(text)

  return text
```

```python
def removeNumber(text) :
  pattern_order = r'[0-9]'
  text = re.sub(pattern_order, '', text)

  return text
```
