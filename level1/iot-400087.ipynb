```python
import numpy as np

#turn time to right and useful format
def time_formatting(df):
  df_copy = df.copy()

  sin_vals = []

  for i in df['time']:
    #print(times)
    time_split = i.split(':')

    hours.append(int(time_split[0]))
    mins.append(int(time_split[1]))
    secs.append(float(time_split[2].split('+')[0]))

    all_time = (int(time_split[0])*3600) + (int(time_split[1])*60) + float(time_split[2].split('+')[0])
    all_time /= 86400
    sin = np.sin(2*np.pi*all_time)

    sin_vals.append(sin)

  df_copy['sin_value'] = sin_vals

  df_copy.drop('time',axis='columns', inplace=True)

  return df_copy
```

```python
from sklearn.preprocessing import MinMaxScaler

def normalizing(df):
  df_copy = df.copy()
  scaler = MinMaxScaler()
  scaler.fit(df_copy)
  for i in df.columns:
    arr = np.array(df_copy[i]).reshape(-1, 1)
    final_normalize = scaler.transform(arr)

    df_copy[i]=final_normalize

  return df_copy
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesRegressor

seed = 123
X = train_df[['1Air Humidity','23SoilEC', '23SoilpH', '26P', '26K', '23SoilTemp.อุณหภูมิดิน','29SoilTension']]
y = train_df[['23SoilWC ค.ชื้นดิน']]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=seed)
reg = ExtraTreesRegressor(n_estimators = 300, random_state=seed, max_depth=30, min_samples_leaf=1, 
                          max_features=1.0, n_jobs=-1
                          ,criterion ='absolute_error').fit(
   X_train, y_train)
reg.score(X_test, y_test)
```
